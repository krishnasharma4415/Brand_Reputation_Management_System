```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train_sample.csv')
```

```python
df.info()
```

```python
df_A = df[['Choice'] + [col for col in df.columns if col.startswith('A_')]].copy()
df_B = df[['Choice'] + [col for col in df.columns if col.startswith('B_')]].copy()
```

```python
df_A.columns = ['Choice'] + [col.replace('A_', '') for col in df_A.columns if col != 'Choice']
df_B.columns = ['Choice'] + [col.replace('B_', '') for col in df_B.columns if col != 'Choice']
```

```python
df_A['influencer'] = df_A['Choice'].apply(lambda x: 1 if x == 0 else 0)
df_B['influencer'] = df_B['Choice'].apply(lambda x: 0 if x == 0 else 1)
```

```python
df_A.drop(columns=['Choice'], inplace=True)
df_B.drop(columns=['Choice'], inplace=True)
```

```python
final_df = pd.concat([df_A, df_B], ignore_index=True)
```

```python
final_df.to_csv('train_influencers.csv', index=False)
```
