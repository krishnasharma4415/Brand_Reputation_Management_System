```python
import pandas as pd
import os

def merge_datasets(folder_path):
    merged_data = {'instagram': [], 'tiktok': [], 'youtube': []}

    for filename in os.listdir(folder_path):
        if filename.endswith('.csv'):
            file_path = os.path.join(folder_path, filename)
            df = pd.read_csv(file_path)

            file_type = filename.split()[0].lower()
            merged_data[file_type].append(df)

    for file_type, dfs in merged_data.items():
        merged_data[file_type] = pd.concat(dfs, ignore_index=True)

    return merged_data
```

```python
folder_path = "K:\Computer Science\AIMLDL\Brand_Reputation_Management_System\Data\social_media"
merged_datasets = merge_datasets(folder_path)
```

```python
instagram_df = merged_datasets['instagram']
tiktok_df = merged_datasets['tiktok']
youtube_df = merged_datasets['youtube']
```
